3. Real Time Detections

```python
cap=cv2.VideoCapture(0)
while True:
    isTrue,frame=cap.read()
    results=model(frame)
    cv2.imshow('Video',np.squeeze(results.render()))
    if cv2.waitKey(20) & 0xFF==ord('z'):
        break

cap.release()
cv2.destroyAllWindows()
```

4. Custom Training

```python
import uuid
import os
import time
```

```python
IMAGE_PATH=os.path.join('data','images')
```

```python
labels=['awake','drowsy']
```

```python
img_num=20
```

```python
cap=cv2.VideoCapture(0)
for label in labels:
    print("Capturing images for label {}".format(label))
    time.sleep(5)
    for img in range(img_num):
        print("Capturing images for label {},img no. {}".format(label,img))
        res,frame=cap.read()
        img_name=os.path.join(IMAGE_PATH,label+'.'+str(uuid.uuid1())+'.jpg')
        cv2.imwrite(img_name,frame)
        cv2.imshow('Image Window',frame)
        time.sleep(2)
        if cv2.waitKey(20) & 0xFF==ord('z'):
            break
cap.release()
cv2.destroyAllWindows()
```
